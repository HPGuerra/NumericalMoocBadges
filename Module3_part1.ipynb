```python
import numpy
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib import animation
from JSAnimation.IPython_display import display_animation
```

```python
def u_initial(nx):
    """Define the initial velocity field with a step function"""
    u_i=numpy.zeros(nx)
    u_i[:(nx/2)]=1
    u_i[(nx/2+1):]=0
    return u_i
```

```python
def computeF(u):
    "Compute the flux based on the velocity field u"
    return (u**2)/2
```

```python
def maccomarck(u,nt,dt,dx):
    un=numpy.zeros((nt,len(u)))
    un[:]=u.copy()
    ustar=u.copy()
    for n in range (1,nt):
        F=computeF(u)
        ustar[1:-1]=u[1:-1]-dt/dx*(F[2:]-F[1:-1])+eps*(u[2:]-2*u[1:-1]+u[:-2])     "The modification is here"
        Fstar=computeF(ustar)
        un[n,1:]=0.5*(u[1:]+ustar[1:]-dt/dx*(Fstar[1:]-Fstar[:-1]))
        u=un[n].copy()
    return un
```

```python
nx=81
nt=70
dx=4.0/(nx-1)
eps=0.5
```

```python
def animate(data):
    x=numpy.linspace(0,4,nx)
    y=data
    line.set_data(x,y)
    return line
```

```python
u=u_initial(nx)
sigma=0.5
dt=sigma*dx
```

```python
un=maccomarck(u,nt,dt,dx)
```

```python
fig=plt.figure()
ax=plt.axes(xlim=(0,4),ylim=(-.5,2))
line,=ax.plot([],[],lw=2)
anim = animation.FuncAnimation(fig, animate, frames=un, interval=50)
display_animation(anim, default_mode='once')
```
